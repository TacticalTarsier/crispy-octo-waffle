```python
import pandas as pd
from fbprophet import Prophet

from prophet.plot import plot_plotly, plot_components_plotly
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/facebook/prophet/main/examples/example_wp_log_peyton_manning.csv')
df = df.astype({'ds': 'datetime64[ns]'})
```

Simple model

```python
pred_model = Prophet()
pred_model.fit(df)
```

```python
future = pred_model.make_future_dataframe(periods=365, freq='D')
```

```python
forecast = pred_model.predict(future)
```

```python
plot_plotly(pred_model, forecast)
```

```python
plot_components_plotly(pred_model, forecast)
```
