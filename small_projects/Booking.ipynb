```python
# Read data

bookings = pd.read_csv('2_bookings.csv', sep=';')
bookings_head = bookings.head(7)
```

Rename columns to make it easier to work with data.

```python
# Create dict consists of lower names and replace spaces by '_'

col = bookings.columns

d_of_col = {}
for i in list(col):
    d_of_col[i] = i.lower().replace(' ', '_')
```

```python
bookings.rename(columns=d_of_col, inplace=True)
```

From what country the most succecsul bookings were done?

```python
bookings.query("is_canceled == 0") \
        .groupby('country', as_index=False) \
        .aggregate('count') \
        .sort_values('hotel', ascending=False) \
        .country \
        .head()
```

On how many nights do they stay in a hotel type 'city hotel' and 'resort hotel' in average? 

```python
round(bookings.query("hotel == 'City Hotel'") \
        .stays_total_nights \
        .mean(), 2)
```

```python
round(bookings.query("hotel != 'City Hotel'") \
        .stays_total_nights \
        .mean(), 2)
```

How many overbooking cases are in data?

```python
bookings.query('assigned_room_type != reserved_room_type') \
        .shape[0]
```

Now lets analyze arrival date. What is the common month to book a hotel in 2016?

```python
bookings.query('arrival_date_year == 2016') \
        .groupby('arrival_date_month', as_index=False) \
        .aggregate('count') \
        .sort_values('hotel', ascending=False)
```

In 2017?

```python
bookings.query('arrival_date_year == 2017') \
        .groupby('arrival_date_month') \
        .aggregate('count') \
        .sort_values('hotel', ascending=False)
```

On which month do they cancel booking the most?

```python
bookings.query('is_canceled == 1') \
        .groupby(['arrival_date_year', 'arrival_date_month'], as_index=False) \
        .aggregate({'is_canceled': 'count'}) \
        .sort_values('is_canceled', ascending=False)
```

Look at columns 'adults', 'children' and 'babies'. Which column does have the biggest avg?

```python
bookings[['adults', 'children', 'babies']] \
        .describe()
```

Create columng 'total_kids' as sum of 'children' and 'babies'. 
For what type an avg value of total_kids is the biggest?

```python
bookings['total_kids'] = bookings['children'] + bookings['babies']
```

```python
bookings.groupby('hotel', as_index=False) \
        .aggregate({'total_kids': 'mean'}) \
        .rename(columns={'total_kids': 'avg_total_kids'})
```

Lets calculate CR (Chunk rate) among peolpe with or without kids

```python
bookings['has_kids'] = bookings.total_kids >= 1
```

```python
bookings.groupby('has_kids', as_index=False) \
        .aggregate({'is_canceled': 'mean'}) \
        .rename(columns={'is_canceled': 'CR'})
```
